Pair Plot:<br>
&emsp;&emsp;&emsp;A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will be shared in the y-axis across a single row and in the x-axis across a single column.It only plots the numerical data.

```python
sns.pairplot(df)
```

```python
# hue parameter determines which column in the data frame should be used for colour encoding
# Adding `hue="smoker" tells seaborn that to colour the data points for smoker and non-smoker differently
sns.pairplot(df, hue='smoker')
```

```python
# Highlighting the day points
sns.pairplot(df, hue='day')
```

```python
# Highlighiting whether points are male or female
sns.pairplot(df, hue='sex')
```
