# **Ordinal Encoding**

##Count/Frequency Method

```python
df = data[['X1','X2']]
```

```python
df.head()
```

```python
print(len(df['X1'].unique()))
print(len(df['X2'].unique()))
```

```python
di = df['X2'].value_counts().to_dict()
di
```

```python
df_count = df['X2'].value_counts().to_dict()
len(df_count)
```

```python
# Mapping the counts/frequency to the categotical variables in the dataset

df.X2 = df.X2.map(df_count)
df.head(10)
```

```python
df.head(10)
```

```python
df_count_X1 = df['X1'].value_counts().to_dict()
df_count_X1
```

```python
df.X1 = df.X1.map(df_count_X1)
```

```python
df.head()
```

##***Label Encoding***

```python
import pandas as pd
import datetime 
```

```python
df_b = datetime.datetime.today()
#List of 20days difference from today
df_date_li = [df_b - datetime.timedelta(days=x) for x in range(0,20)]
df1 = pd.DataFrame(df_date_li)
df1.columns = ['day']
df1.head()
```

```python
# extract the week day name

df1['day_of_week'] = df1['day'].dt.day_name()
df1.head()
```

```python
df1.head()
```

```python
df1_weeks = {'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
df1['label_encoded'] = df1.day_of_week.map(df1_weeks)
```
